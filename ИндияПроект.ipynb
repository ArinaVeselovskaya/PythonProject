**1.1. Оценка динамики выдачи рабочих карточек и связь с ростом населения Индии**
Глобальная оценка динамики выдачи рабочих карточек показывает, что начиная с 2017-18 экономичекого года количество выданных карточек стремительно росло, что говорит об актуальности и востребованности программы в Индии.

```python
# @title
x = []
y = []

al_a = data[["Year", "Job cards issued"]]
for i in sorted(list(set(data["Year"]))):
  df = al_a[al_a["Year"] == i]
  a = sum(df["Job cards issued"])
  x.append(i)
  y.append(a/1000000)
  #print(a)
#print(y)
#print(x)
```

```python
# @title
f = plt.figure()
f.set_figwidth(10)  #изменяем длину оси х , затем у
f.set_figheight(4)

plt.xlabel('Экономический год') #Подпись для оси х
plt.ylabel('Кол-во выданных карточек, млн') #Подпись для оси y
plt.title('Количество выданных карточек')
plt.grid(color='#BABABA')

plt.plot(x, y, color='green', marker='o', markersize=6)  #строит график

plt.show()
```

Давайте попробуем разобраться, с чем может быть связан рост количества выданных карточек.

Для начала оценим взаимосвязь с населением Индии.

```python
# @title
populat = []
al_a = population['Население']

for i in list(population["Население"]):
  populat.append(i)

# print(populat)

df = {
    "Население Индии": populat,
    "Количество выданных карточек": y
}

data = pd.DataFrame(df)

print(data.corr())
```

Коэффициент корреляции равен **0.533073**, что говорит об умеренной прямопропорциональной зависимости роста населения Индии и количества выданных карточек.

**1.2.   Оценка динамики затрат на оплату труда и общего количества отработанных человеко-дней**
Для начала посмотрим на количество денег, потраченных на оплату труда и количество отработанных человеко-дней. Логично предположить, что из роста количества выданных карточек следует рост количества отработанных человеко-дней и количества денег, потраченных на зарплаты рабочим.

```python
# @title
data = pd.read_excel("IndiaPython.xlsx")

x = []
money = []
human_h =[]
medium_mon = []
al_a = data[["Year", "Labour expenditure that has been disbursed", "Total person days"]]

for i in sorted(list(set(data["Year"]))):
  df = al_a[al_a["Year"] == i]
  df['Labour expenditure that has been disbursed'] = df['Labour expenditure that has been disbursed']. fillna (0)
  len_mon = len(df["Labour expenditure that has been disbursed"])
  a_mon = sum(df["Labour expenditure that has been disbursed"])
  a_hum = sum(df["Total person days"])
  x.append(i)
  medium_mon.append(a_mon/len_mon)
  money.append(a_mon/1000000) # в млн
  human_h.append(a_hum/1000000) #  в млн



# print(money)
# print(human_h)
# print(medium_mon)
```

```python
# @title
f = plt.figure()
f.set_figwidth(18)  #изменяем длину оси х , затем у
f.set_figheight(4)

plt.subplot(121)
plt.xlabel('Экономический год') #Подпись для оси х
plt.ylabel('Деньги, потраченные на оплату труда, млн $') #Подпись для оси y
plt.title('Количество денег, потраченных на оплату труда')
plt.grid(color='#BABABA')

plt.plot(x, money, color='#CD4F39', marker='o', markersize=6)  #строит график

plt.subplot(122)
plt.xlabel('Экономический год') #Подпись для оси х
plt.ylabel('Человеко-дни, млн') #Подпись для оси y
plt.title('Количество отработанных человеко-дней')
plt.grid(color='#BABABA')

plt.plot(x, human_h, color='#008B45', marker='o', markersize=6)  #строит график

plt.show()
```

На графиках мы видим прямую зависимость уже перечисленных параметров. Однако количесвто выделенных средств на зарплаты, как и количество отработанных человек-дней, начинает расти с 2014-15 экономического года, а не с 2017-18. Это может говорить о том, что существует множество других факторов, влияющих на рост востребованности программы гражданами Индии.  

**1.3.   Оценка средней оплаты человеко-дня за все года и связь с ВВП Индии**

Одним из таких факторов может является повышение средней оплаты человека-дня: чем больше платят за работу, тем больше желающих появится получить эту работу. Проверим гипотезу.

```python
# @title
f = plt.figure()
f.set_figwidth(16)  #изменяем длину оси х , затем у
f.set_figheight(4)

plt.xlabel('Экономический год') #Подпись для оси х
plt.ylabel('Средняя оплата за год, $') #Подпись для оси y
plt.title('Средняя оплата за год по всем штатам')
plt.grid(color='#BABABA')

plt.plot(x, medium_mon, color='#CD4F39', marker='o', markersize=6)  #строит график
```

На графике видно, что средняя оплата человеко-дня росла почти каждый год, что может служить подтверждением выдвинутой гипотезы.

Выясним, как связаны ВВП Индии и средняя оплата человеко-дня по программе MGNREGA.

```python
# @title
vvp = []

for i in list(VVP["ВВП(млрд)"]):
  vvp.append(i)

# print(vvp)

df = {
    "ВВП Индии": vvp,
    "Средняя оплата человеко-дня": medium_mon
}

data = pd.DataFrame(df)

print(data.corr())
```

Коэффициент корреляции равен 0.919, это показатель того, что ВВП Индии и средняя оплата человек-дня по программе MGNREGA имеют сильную прямопропрорциональную связь. Иными словами, можно сказать, что процветание внутренней экономики Индии позволило увеличить среднюю оплату труда, а увеличение оплаты труда повысило спрос на участие в программе MGNREGA, что позволило закрепить повышение ВВП страны.

**1.4.   Оценка количества выплаченных денег трем бедным и трем богатым штатам Индии**

А теперь посмотрим на те же данные, но локально.
Сконцентрируем внимание на трех бедных штатах Индии: Bihar, Jharkhand, Uttar Pradesh; и на трех богатых штатах Индии: Kerala, Goa, Sikkim.
Оценим динамику количества выплаченных денег вышеперечисленным штатам.

```python
# @title
year = []
poor = []
rich = []
Bihar_y = []
Jharkhand_y = []
UttarPradesh_y = []
Sikkim_y = []
Goa_y = []
Kerala_y = []
all_money = []

data = pd.read_excel("IndiaPython.xlsx")
al_a = data[["Year", "State", "Labour expenditure that has been disbursed"]]
for i in sorted(list(set(data["Year"]))):
  df_Bihar = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Bihar')]
  df_Bihar['Labour expenditure that has been disbursed'] = df_Bihar['Labour expenditure that has been disbursed']. fillna (0)
  Bihar_sum = sum(df_Bihar["Labour expenditure that has been disbursed"])
  Bihar_y.append(Bihar_sum/1000)
  year.append(i)
  #print(Bihar_sum)


for i in sorted(list(set(data["Year"]))):
  df_Jharkhand = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Jharkhand')]
  df_Jharkhand['Labour expenditure that has been disbursed'] = df_Jharkhand['Labour expenditure that has been disbursed']. fillna (0)
  Jharkhand_sum = sum(df_Jharkhand["Labour expenditure that has been disbursed"])
  Jharkhand_y.append(Jharkhand_sum/1000)
  #print(Jharkhand_sum)

for i in sorted(list(set(data["Year"]))):
  df_UttarPradesh = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Uttar Pradesh')]
  df_UttarPradesh['Labour expenditure that has been disbursed'] = df_UttarPradesh['Labour expenditure that has been disbursed']. fillna (0)
  UttarPradesh_sum = sum(df_UttarPradesh["Labour expenditure that has been disbursed"])
  UttarPradesh_y.append(UttarPradesh_sum/1000)
  #print(UttarPradesh_sum)

# к самому богатому
for i in sorted(list(set(data["Year"]))):
  df_Sikkim = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Sikkim')]
  df_Sikkim['Labour expenditure that has been disbursed'] = df_Sikkim['Labour expenditure that has been disbursed']. fillna (0)
  Sikkim_sum = sum(df_Sikkim["Labour expenditure that has been disbursed"])
  Sikkim_y.append(Sikkim_sum/1000)
  #print(Sikkim_sum)

for i in sorted(list(set(data["Year"]))):
  df_Goa = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Goa')]
  df_Goa['Labour expenditure that has been disbursed'] = df_Goa['Labour expenditure that has been disbursed']. fillna (0)
  Goa_sum = sum(df_Goa["Labour expenditure that has been disbursed"])
  Goa_y.append(Goa_sum/1000)
  #print(Goa_sum)

for i in sorted(list(set(data["Year"]))):
  df_Kerala = al_a.loc[(al_a["Year"] == i) & (al_a["State"] == 'Kerala')]
  df_Kerala['Labour expenditure that has been disbursed'] = df_Kerala['Labour expenditure that has been disbursed']. fillna (0)
  Kerala_sum = sum(df_Kerala["Labour expenditure that has been disbursed"])
  Kerala_y.append(Kerala_sum/1000)
  #print(Kerala_sum)



poor = list(map(sum, zip(Bihar_y,Jharkhand_y,UttarPradesh_y)))
rich = list(map(sum, zip(Sikkim_y,Goa_y,Kerala_y)))



for i in sorted(list(set(data["Year"]))):
  df = al_a[al_a["Year"] == i]
  df['Labour expenditure that has been disbursed'] = df['Labour expenditure that has been disbursed']. fillna (0)
  money = sum(df["Labour expenditure that has been disbursed"])
  all_money.append(money/1000)


# print(poor)
# print(rich)
# #print(rich)
# #print(Bihar_y)
# #print(Jharkhand_y)
# #print(UttarPradesh_y)
# #print(Sikkim_y)
# #print(Goa_y)
# #print(Kerala_y)
# print(all_money)
```

```python
# @title
f = plt.figure()
f.set_figwidth(10)  #изменяем длину оси х , затем у
f.set_figheight(4)

plt.xlabel('Экономический год') #Подпись для оси х
plt.ylabel('Выплаченные деньги, тыс $') #Подпись для оси y
plt.title('Выплаченные деньги трём самым богатым и бедным штатам')
plt.grid(color='#BABABA')

plt.plot(year, poor, color='#473C8B', marker='o', markersize=6)  #строит график
plt.plot(year, rich, color='#EE7942', marker='o', markersize=6)
plt.legend(['Bihar, Jharkhand, Uttar Pradesh', 'Sikkim, Goa, Kerala'])

plt.show()
```

3.2. **Оценка зависимости роста количества заявок от домохозяйств от показателя ВВП Индии.**

Для оценки зависимости будем использовать метод квадратов Пирсона. Коэффициент, полученный при рассчетах, покажет степень зависимость.

```python
# @title
# I
ari_zayavki = []
ari_ab = data[["Year", "Households that applied for a job card"]]
for i in sorted(list(set(data["Year"]))):
  df_123_ari = ari_ab[ari_ab["Year"] == i]
  applied_1 = sum(df_123_ari["Households that applied for a job card"])
  ari_zayavki.append(applied_1)

srednie_zayavki = (sum(ari_zayavki)/len(ari_zayavki))

for i in range(len(ari_zayavki)):
  ari_zayavki[i] = ari_zayavki[i] - srednie_zayavki


# II
ari_vvp = []
ari_ac = VVP[["Год", "ВВП(млрд)"]]
for i in sorted(list(set(VVP["Год"]))):
  df_124_ari = ari_ac[ari_ac["Год"] == i]
  ari_vvp_chislo = sum(df_124_ari["ВВП(млрд)"])
  ari_vvp.append(ari_vvp_chislo)

sravni_vvp = sum(ari_vvp)/len(ari_vvp)

for i in range(len(ari_vvp)):
  ari_vvp[i] = ari_vvp[i] - sravni_vvp


# Рассчитываем коэфф
ari_verh_0 = 0

for i in range(len(ari_vvp)):
  ari_verh_0 += ari_vvp[i] * ari_zayavki[i]

ari_pro_1 = 0
for i in ari_vvp:
  ari_pro_1 += i**2

ari_pro_2 = 0
for i in ari_zayavki:
  ari_pro_2 += i**2

ari_niz_0 = (ari_pro_1 * ari_pro_2)**0.5

Rr = ari_verh_0/ari_niz_0
print('Коэффициент корреляции Пирсона =', Rr)
```

Значение коэффициента в промежутке от 0,3 до 0,69 указывает на наличие средней прямой зависимости.

Значит, с увеличением ВВП Индии увеличивалось и количество заявок от домохозяйств.

3.7. **Оценка зависимости между уровнем бедности и долей одобренных заявок.**

Для оценки зависимости будем использовать метод квадратов Пирсона. Коэффициент, полученный при рассчетах, покажет, какова зависимость.

```python
# @title
dolya_prinyat = []
index_bednosty = []
sredni_dolya = []
sredni_index = []
# первая часть
ari_a2 = data[["State", "Households that were allotted work", "Households that applied for a job card"]]
for i in sorted(list(set(data["State"]))):
  if i != 'Ladakh':
    df2_state = ari_a2[ari_a2["State"] == i]
    ari_applied = sum(df2_state["Households that applied for a job card"])
    ari_allotted = sum(df2_state["Households that were allotted work"])
    dolya_prinyat_chislo = (ari_allotted/ari_applied) * 100
    dolya_prinyat.append(dolya_prinyat_chislo)

sredni_dolya_chislo = sum(dolya_prinyat)/len(dolya_prinyat)

for i in dolya_prinyat:
  sredni_dolya.append(i - sredni_dolya_chislo)    #список с долями после вычета




# вторая часть
ari_a3 = poverty[["Штат", "Индекс бедности"]]
for i in sorted(list(set(poverty["Штат"]))):
  df3_state = ari_a3[ari_a3["Штат"] == i]
  index_bednosty_chislo = sum(df3_state["Индекс бедности"])
  index_bednosty.append(index_bednosty_chislo)

sredni_index_chislo = sum(index_bednosty)/len(index_bednosty)

for i in index_bednosty:
  sredni_index.append(i - sredni_index_chislo)    #список с индексами после вычета


# счтаем коэф
ari_rez_verh = 0
for i in range(len(sredni_index)):
  ari_1 = sredni_index[i] * sredni_dolya[i]
  ari_rez_verh += ari_1

ari_2 = 0
for i in sredni_dolya:
  ari_2 += i**2

ari_3 = 0
for i in sredni_index:
  ari_3 += i**2

ari_rez_niz = (ari_2 * ari_3) ** 0.5

R_1 = ari_rez_verh/ari_rez_niz

print("Коэффициент корреляции Пирсона =", R_1)
```

При таком значении коэффициента можно утверждать, что между долей одобренных заявок от домохозяйств и уровнем бедности штатов зависимости нет.

```python
# @title
                             # Достаем данные
ari1_dict = {}
ari1_worked_households = []
ari1_m = data[["Households that worked under mahatma gandhi national rural employment guarantee act (mgnrega)", "State", "Households that were allotted work", "Households that applied for a job card"]]
# print(list(set(data["State"])))
# print(len(list(set(data["State"]))))
for i in list(set(data["State"])):
  if i != 'Ladakh':
    df1_state = ari_m[ari_m["State"] == i]
    ari1_households_allotted = sum(df1_state["Households that were allotted work"])
    ari1_households_worked = sum(df1_state["Households that worked under mahatma gandhi national rural employment guarantee act (mgnrega)"])
    ari1_sravnen = round(((ari1_households_worked/ari1_households_allotted)*100)//1)
    ari1_dict[i] = ari1_sravnen
  # print(ari_dict)
  sort1_data = sorted(ari1_dict.items(), key=lambda x: x[1], reverse=True)
  sort1_ari_dict = dict(sort1_data)
  # print(sort_ari_dict)
  ari1_keys = list(sort1_ari_dict.keys ())  #список штатов
# print(ari1_dict)
# print(ari1_keys)
for i in (ari1_keys):
  ari1_worked_households.append(sort1_ari_dict[i])
# print(ari1_worked_households)  #список доли выполненной работы.

                       # Рассчет коэффициента:
srednaya_dolya = sum(ari1_worked_households)/len(ari1_worked_households)
# print(srednaya_dolya)
otklonenie_ot_sred_doli = []    #отклонение от средней доли, начиная с самой большой
for i in (ari1_worked_households):
  otklonenie_ot_sred_doli.append(i - srednaya_dolya)
# print(otklonenie_ot_sred_doli)



bednost = list(poverty["Индекс бедности"])
otklonenie_ot_sred_bednost = []
srednaya_bednost = sum(bednost)/len(bednost)
for i in (bednost):
  otklonenie_ot_sred_bednost.append(i - srednaya_bednost)
# print(otklonenie_ot_sred_bednost)

sum_verh = 0
for i in range(len(otklonenie_ot_sred_bednost)):
  ari_proizv1 = otklonenie_ot_sred_bednost[i] * otklonenie_ot_sred_doli[i]
  sum_verh += ari_proizv1

aa = 0
for i in otklonenie_ot_sred_bednost:
  aa += i ** 2
aa = aa ** 0.5

bb = 0
for i in otklonenie_ot_sred_doli:
  bb += i ** 2
bb = bb ** 0.5
cc = aa * bb

R = sum_verh/cc

print("Коэффициент корреляции Пирсона =", R)
```

Значение коэффициента в промежутке от 0 до –0,29 указывает на наличие слабой обратной зависимости.

То есть, чем больше индекс бедности, тем меньше доля выполненной работы, но зависимость слабая.
